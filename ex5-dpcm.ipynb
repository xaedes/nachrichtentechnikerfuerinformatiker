```python
def f_signal(ts,freq=5):
    return np.sin(2*np.cos(math.pi*2*freq*ts))
```

```python
def f_signal2(ts,lin=1,freq=2,freq2=3,freq3=3,t1=0,t2=0,t3=0):
    return ts*lin + (1+np.sin(math.pi*2*freq*(ts+t1)))*(1+np.sin(math.pi*2*freq2*(ts+t2)))*(1+np.sin(math.pi*2*freq3*(ts+t3)))
```
