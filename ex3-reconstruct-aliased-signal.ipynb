```python
def fft_reconstruct(f,ts):
    dt = np.diff(ts).mean()
    f_spectrum = np.fft.fft(f(ts))
    reconstructed = np.fft.ifft(f_spectrum)
    return reconstructed
```

```python
ts = np.linspace(0,1,200)
plot(ts,f(ts,freq=40))
plot(ts,fft_reconstruct(lambda ts:f(ts,freq=40),ts))
```
