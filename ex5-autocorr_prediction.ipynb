```python
%pylab inline
from pprint import pprint
from funcy import partial # sudo pip install funcy
from autocorrelate import autocorrelate
from autocorr_prediction import build_matrix, predict, predict_n
```

```python
def f_signal(ts,freq=5):
    return np.sin(math.pi*2*freq*ts)
```

```python
def f_signal2(ts,lin=1,freq=2,freq2=3,freq3=3,t1=0,t2=0,t3=0):
    return ts*lin + (1+np.sin(math.pi*2*freq*(ts+t1)))*(1+np.sin(math.pi*2*freq2*(ts+t2)))*(1+np.sin(math.pi*2*freq3*(ts+t3)))
```

```python
n = 100
n_predict = 100
ts = np.linspace(0,1,n)
signal = f_signal(ts)
signal = f_signal2(ts,0,2,20,30,0,1,3)

signal -= np.mean(signal)
```

```python
plot(ts,signal,"b")
plot(ts,signal,"bx")
```

```python
predict(signal)
```

```python
signal_predictions = dict()
types = [0]
for autocorr_type in types:
    signal_prediction = predict_n(signal, n_predict=100, time_horizon=100,autocorr_type=autocorr_type)
    signal_predictions[autocorr_type] = signal_prediction
```

```python
for autocorr_type,signal_prediction in signal_predictions.iteritems():
    figure(figsize=(10,5))
    plot(signal_prediction,"r")
    plot(signal,"b")
    ylim(np.min(signal)-.1,np.max(signal)+.1)
    title(autocorr_type)
```
