```python
%pylab inline
from pprint import pprint
from funcy import partial # sudo pip install funcy
from autocorrelate import autocorrelate
from autocorr_prediction import predict, predict_n
```

```python
signal = np.loadtxt("eurokurs.csv")
n=len(signal)
#signal=np.cumsum(np.random.normal(0,1,n))
signal = np.diff(signal)
signal = signal - np.mean(signal)
plt.plot(signal)
```

```python
n_predict = 100
```

```python
signal_prediction = predict_n(signal, n_predict, time_horizon=200)
```

```python
plot(signal_prediction,"r")
#plot(signal_prediction,"rx")
plot(signal,"b")
#plot(signal,"bx")
```

```python
plot(np.cumsum(signal_prediction),"r")
plot(np.cumsum(signal),"b")
```

```python
for t in range(200,1200,200):
    figure()
    prediction = predict_n(signal[:t], n_predict=200, time_horizon=200)
    plot(np.cumsum(signal),"b")
    plot(np.cumsum(prediction),"r")
    plot(np.cumsum(signal[:t]),"b")
```
