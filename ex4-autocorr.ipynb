```python
def autocorrelate(xs,tau):
    if tau == 0:
        s1 = xs
        s2 = xs
    elif tau > 0:
        s1 = xs[:-tau]
        s2 = xs[tau:]
    elif tau < 0:
        s1 = xs[:tau]
        s2 = xs[-tau:]
        
    mu = np.mean(xs)
    mu1 = np.mean(s1)
    mu2 = np.mean(s2)
    var = np.std(xs)**2
    
    #return np.mean((s1-mu1)*(s2-mu2))/var
    #return np.mean((s1-mu1)*(s2-mu2))
    #return np.mean((s1-mu)*(s2-mu))/var
    #return np.mean((s1-mu)*(s2-mu))
    #return np.mean(s1*s2)/var
    #return np.sum(s1*s2)/var
    return np.sum(s1*s2)
```
