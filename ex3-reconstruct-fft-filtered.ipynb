```python
%pylab inline
from funcy import partial
from collections import OrderedDict
import scipy.signal
```

```python
def f(t,freq=100):
    x = np.sin(2*math.pi*freq*t)
    return x
```

```python
def f_multi(t,max_freq=100):
    x = np.zeros_like(t)
    for freq in range(max_freq):
        x += np.sin(2*math.pi*freq*t)
    return x
```

```python
n = 200
ts = np.linspace(0,1,n)
dt = np.diff(ts).mean()
xs = f_multi(ts,50)
plt.plot(ts,xs)
```

```python
def fft_reconstruct(f,ts):
    dt = np.diff(ts).mean()
    f_spectrum = np.fft.fft(f(ts))
    reconstructed = np.fft.ifft(f_spectrum)
    return reconstructed
```

```python
def fft_filter(f,ts,filter_function):
    dt = np.diff(ts).mean()
    f_spectrum = np.fft.fft(f(ts))
    fftfreq = np.fft.fftfreq(ts.shape[-1],dt)
    f_spectrum = filter_function(fftfreq,f_spectrum)
    reconstructed = np.fft.ifft(f_spectrum)
    return reconstructed
```

```python
def perfect_low_pass(cutoff,freq,spectrum):
    spectrum[np.abs(freq) < cutoff] = 0
    return spectrum
    
```

```python
def perfect_high_pass(cutoff,freq,spectrum):
    spectrum[np.abs(freq) > cutoff] = 0
    return spectrum
    
```

```python
def perfect_band_pass(minfreq,maxfreq,freq,spectrum):
    spectrum[np.logical_or(freq < minfreq,freq>maxfreq)] = 0
    return spectrum
    
```

```python
ts = np.linspace(0,1,1000)
max_freq = 100
cutoff = 20
fun = lambda ts: f_multi(ts,max_freq)
plt.figure(figsize=(10,7))
plot(ts,fun(ts))
plot(ts,fft_filter(
    f = fun, ts = ts,
    filter_function = partial(perfect_low_pass,cutoff)))
plot(ts,fft_filter(
    f = fun, ts = ts,
    filter_function = partial(perfect_high_pass,cutoff)))
```
